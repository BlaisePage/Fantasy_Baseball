#### How do we pick our team?

In fantasy baseball, each player can only belong to one team (just like in real life). You will be selecting your team with a partner through an **auction draft**. In an auction draft, you'll start with \\$250. You'll take turns nominating a player, and then everyone will live bid on that player, just like a real life auction. The least a player can go for is \\$1, while the most is \\$250-(# roster spots - 1). #### What are waivers?

Waivers are where any non-rostered player is at. You can pick up players from waivers, but you'll have to drop a player from your roster to do so.

```python
# pitcher selection

import pandas as pd
import numpy as np
from pybaseball import pitching_stats

pitch_stats = pitching_stats(2018)
pitch_stats.head(10)
```

```python
#print(pitch_stats['Relieving'])
list(pitch_stats)
```

```python
#print(len(pitch_stats)) # 784
from collections import OrderedDict

#starting_pitcher_dict = OrderedDict()
#relieving_pitcher_dict = OrderedDict()
pitcher_dict = OrderedDict()

# populate pitcher_dict with pitcher names as keys
for index, row in pitch_stats.iterrows():
    # if pitcher is a starter
    #if(row['Start-IP']>=row['Relief-IP']):
    #    starting_pitcher_dict[row['Name']] = {}
    # if pitcher is a reliever
    #else:
    #    relieving_pitcher_dict[row['Name']] = {}
    
    # make the master key the pitchers name
    pitcher_dict[row['Name']] = OrderedDict()
    # IP data
    pitcher_dict[row['Name']]['IP'] = row['IP']
    # K data
    pitcher_dict[row['Name']]['K'] = row['SO']
    # BB data
    pitcher_dict[row['Name']]['BB'] = row['BB']
    # HBP data
    pitcher_dict[row['Name']]['HBP'] = row['HBP']
    # HR data
    pitcher_dict[row['Name']]['HR'] = row['HR']
    # SV data
    pitcher_dict[row['Name']]['SV'] = row['SV']
    # HLD data
    pitcher_dict[row['Name']]['HLD'] = row['HLD']
    # WAR
    pitcher_dict[row['Name']]['WAR'] = row['WAR']
    
#print(pitcher_dict)
#for pitcher in pitcher_dict:
#    print(pitcher_dict[pitcher])
```

```python
# Calculate each pitchers fantasy score as described in the writeup above
# IP=5; K=2; BB=-3; HBP=-3; HR=-13; SV=5; HLD=4
for pitcher in pitcher_dict:
    pitcher_dict[pitcher]['Fantasy_Score'] = 5*pitcher_dict[pitcher]['IP'] + 2*pitcher_dict[pitcher]['K'] + -3*pitcher_dict[pitcher]['BB'] + -3*pitcher_dict[pitcher]['HBP'] + -13*pitcher_dict[pitcher]['HR'] + 5*pitcher_dict[pitcher]['SV'] + 4*pitcher_dict[pitcher]['HLD']
    
#print(pitcher_dict)
```

```python
# Calculate final score based on applying weights to both the Fantasy_Score and WAR
Fantasy_weight = .05
WAR_weight = 3.185

sum_scores = 0

for pitcher in pitcher_dict:
    pitcher_dict[pitcher]['Final_Score'] = Fantasy_weight*pitcher_dict[pitcher]['Fantasy_Score'] + WAR_weight*pitcher_dict[pitcher]['WAR']
    sum_scores += pitcher_dict[pitcher]['Final_Score']
    #print(pitcher, ":",pitcher_dict[pitcher])
    
avg_scores = sum_scores/len(pitch_stats)
print(avg_scores)
```

### Introduction ###
Fantasy sports have become increasingly popular over the last decade with more and more participants becoming involved each year. With more players and larger leagues, it is difficult for individual participants to gain a competitive advantage over their adversaries using personally acquired knowledge alone. That is, routinely watching sports alone does not give fantasy sports competitors enough of an advantage to build reliable fantasy teams that consistantly out perform their opponents.  

For this assignment, we developed a set of algorithms that we believe will help us draft fantasy baseball players that will maximize our teams performance within our league. In other words, our algorithms will help us draft the ideal players for out playing our league opponents. We split these algorithms into two unique sets - one for pitchers and one for batters. 

### Data and Statistics ###
The pitching algorithms are relatively simple. We first took data from the 2018 MLB season (since this was the most recent season to present day). We then determined the fantasy score for all pitchers in this dataset. Note that a pitcher's fantasy score depended on the following statistics: 
* Innings Pitched (IP) = 5 points each
* Strikeout (K) = 2 points each
* Walk (BB) = -3 points each
* Hit By Pitch (HBP) = -3 points each
* Homerun (HR) = -13 points each
* Save (SV) = 5 points each
* Hold (HLD) = 4 points each

Once we determined each pitchers fantasy score, we then determined each pitchers Wins Above Replacement (WAR) metric using the Fangraphs evaluation for WAR. We believed that WAR was the most significant statistic for choosing a fantasy pitcher because it compares a given pitchers stats to that of an average pitcher. By fault, the pitchers with higher WARs can be viewed as better pitchers compared to their league average replacements. We then determined the pitchers' final scores by taking a weighted sum of their WAR values and their fantasy scores. Note that we set up the weight proportions so that the average final score equaled the average amount of money we had to spend for each player - \$13.88. With the weights set up this way, our pitcher selection algorithm not only determines a score for a pitcher, but it also determines a maximum amount of money to spend on said player.

Similarly, for the batting algorithms we took the data from the 2018 MLB season and used that to create the fantasy score for all of the offensive players in the dataset. For the batting algorithm we did not use the traditional WAR statisitc as defense does not matter in terms of scoring fantasy points. As it is pointed out in the assignment however, we still are going to want to be able to determine a player's skill relative to their position so that we place higher value on players who score high points from positions that are not as "deep".  A batter's fantasy score depends on the following statistics: 
* At Bats (AB) = -1
* Hits (H) = 5.6
* Doubles (2B) = 2.9
* Triples (3B) = 5.7
* Home Runs (HR) = 9.4
* Walks (BB) = 3
* Hit by Pitch (HBP) = 3
* Stolen Bases (SB) = 1.9
* Caught Stealing (CS) = -2.8

So we took the overall results from the 2018 season and then modify that number based on their relative skill level within the position and weighting to determine the maximum amount of money to spend on the player.

### Limitations ###
One limitation of our pitching metric is that severe "skewedness" of the final metric we use for pitcher evaluation. Since some pitchers acquired massive stats in 2018, their final scores were considerably higher compared to the rest of the average pitchers. Due to this skew in the data, our top pitchers have incredibly large salary caps that could get us into trouble during the draft. That is if we go big on one of these pitchers, we will most likely have to compromise with average pitchers for the remainder of our team. Come crunch time, it may be difficult to determine which average pitchers we want to draft if we have already drafted a big name pitcher.<br>
A limitation on the batting metrics is that stats are going to skew heavily in favor or players who were not injured last season because the majority of the weight comes from their total batting numbers acrossed all of the games. So if a player is out for part of the season due to injury, or was rested or benched for some period of the season that is going to strongly impact the fantasy score that they received with our system.
